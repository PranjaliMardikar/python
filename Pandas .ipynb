```python
df1 = pd.DataFrame({"A" : [1,2,3,4,5] ,
                   "B" : [6,7,8,9,0] , 
                   "C" : [10,11,12,13,14] ,
                   "D" :[15,16,17,18,19]})
```

```python
df1
```

```python
df2 = pd.DataFrame({"A" : [20,21,22,23,24] ,
                   "B" : [25,26,27,18,29] , 
                   "C" : [30,31,32,33,34] ,
                   "D" :[35,36,37,38,39]})
```

```python
df2
```

```python
df3=pd.concat ([df1 ,df2] , axis=0 , ignore_index =True)
df3
```
